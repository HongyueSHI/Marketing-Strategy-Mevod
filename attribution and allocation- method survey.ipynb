```python
channel_count_2019_07 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2019_08 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2019_09 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2019_10 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2019_11 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2019_12 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2020_01 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}
channel_count_2020_02 = {'bing':0, 'display':0, 'facebook':0, 'search':0, 'youtube':0}

channel_list = ['bing','display','facebook','search','youtube']


for item in df_2019_07_dict:
    
    if item['attribution_survey'] in channel_list:
        channel_count_2019_07[item['attribution_survey']] += 1

            
for item in df_2019_08_dict:
    
    if item['attribution_survey'] in channel_list:
        channel_count_2019_08[item['attribution_survey']] += 1
            
            
for item in df_2019_09_dict:
    
    if item['attribution_survey'] in channel_list:
        channel_count_2019_09[item['attribution_survey']] += 1
            
            
for item in df_2019_10_dict:
    
    if item['attribution_survey'] in channel_list:
        channel_count_2019_10[item['attribution_survey']] += 1
            

for item in df_2019_11_dict:
    

    if item['attribution_survey'] in channel_list:
        channel_count_2019_11[item['attribution_survey']] += 1


for item in df_2019_12_dict:
    

    if item['attribution_survey'] in channel_list:
        channel_count_2019_12[item['attribution_survey']] += 1



for item in df_2020_01_dict:
    

    if item['attribution_survey'] in channel_list:
        channel_count_2020_01[item['attribution_survey']] += 1

            
            
for item in df_2020_02_dict:
    

    if item['attribution_survey'] in channel_list:
        channel_count_2020_02[item['attribution_survey']] += 1
```
